```python
import pandas as pd
import numpy as np
import seaborn as sea
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.figure_factory as ff
import warnings
warnings.filterwarnings("ignore")
```

# Exibir a versão de cada lib

```python
def verificaVersão(lib, libModule):
    try:
        print(f"A versão da biblioteca {lib} é {libModule.__version__}")

    except AttributeError:
        try:
            print(f"A versão da biblioteca {lib} é {libModule.version}")
        except AttributeError:
            print(f"A biblioteca {lib} não tem modulo de versão")
    except Exception as e:
        print(f"Houve um erro inesperado -> {e}")

verificaVersão("pd",pd)
verificaVersão("np",np)
verificaVersão("sea",sea)
verificaVersão("plt",plt)
verificaVersão("px",px)
verificaVersão("go",go)
verificaVersão("make_subplots",make_subplots)
verificaVersão("ff",ff)
```

# renomear colunas

```python
#vendo as colunas
display(df.head())

#renomeando

df = df.rename(columns=({
    "gender": "genero",
    "race/ethnicity":"raca/etinia",
    "parental level of education": "nivel_escolaridade_dos_pais",
    "lunch":"almoco",
    "test preparation course":"curso_preparatorio_para_test",
    "math score":"nota_matematica",
    "reading score":"nota_leitura",
    "writing score":"nota_escrita"
    }))

display(df.head())
```

```python
display(df.head())
df = df.replace({"male":"masculino", "female": "feminino"})
display(df.head(3))
```

# filtrar valores unicos

```python
for column in df.columns:
    print(f"Valores unicos na coluna {column}")
    if df[column].dtype == "object":
        print(df[column].unique())
    else:
        print(df[column].nunique())
    print('-' * 50) 
```

# Salvando a nova base com modificações

```python
nomeNovo = "./excel/novo_excel_modificado.xlsx"

if df.to_excel(nomeNovo, index=False) == True:
    print(f"Arquivo salvo na pasta excel com o nome de {nomeNovo}")
else:
    print("O arquivo já existe no destino 'excel'")
```
